```python
response = (
    youtube
    .commentThreads()
    .list(
      part='snippet,replies', 
      videoId='iFnIPPorydo', 
      maxResults=100
      )
    .execute()
    )

items = len(response.get('items'))

comments = {
    'commentor' :  [response.get('items')[i].get('snippet').get('topLevelComment').get('snippet').get('authorDisplayName') for i in range(items)],
    'comments' : [response.get('items')[i].get('snippet').get('topLevelComment').get('snippet').get('textOriginal') for i in range(items)]
}

comments = pd.DataFrame(comments)

while 'nextPageToken' in response:
  response = (
    youtube
    .commentThreads()
    .list(
      part='snippet,replies',
      videoId='iFnIPPorydo',
      pageToken=response.get('nextPageToken'),
      maxResults=100
      )
    .execute()
    ) 
  items = len(response.get('items'))

  comment = {
        'commentor' :  [response.get('items')[i].get('snippet').get('topLevelComment').get('snippet').get('authorDisplayName') for i in range(items)],
        'comments' : [response.get('items')[i].get('snippet').get('topLevelComment').get('snippet').get('textOriginal') for i in range(items)]
    }
  comments = comments.append(pd.DataFrame(comment))
```
